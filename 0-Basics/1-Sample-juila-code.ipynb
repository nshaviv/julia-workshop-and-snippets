Now let's plot a plot of $y = \sin(x)$

```julia

x = 0:0.1:10 # Create a vector of x-values
y = sin.(x); # Create a vector of y-values
```

```julia
plot(x,y,frame=:box,label="sin") # Plot the data    
```
